```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Admission_Prediction.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['GRE Score'].fillna(df['GRE Score'].mean(),inplace=True)
df['TOEFL Score'].fillna(df['TOEFL Score'].mean(),inplace=True)
df['University Rating'].fillna(df['University Rating'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Serial No.'],inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1]
```

```python
y=df['Chance of Admit']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=33)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=svr.predict(x_test)
```

```python
score=r2_score(y_test,y_pred)
```

```python
score
```

## HyperParameter Tunning- Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'C':[0.1,1,10,50,100,500],
    'gamma':[1,0.5,0.1,0.01,0.001]
}
```

```python
grid=GridSearchCV(SVR(), param_grid, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_estimator_
```

```python
new_svr=SVR( C=50, gamma=0.001, cache_size=200, coef0=0.0, degree=3,
            epsilon=0.1, kernel='rbf', max_iter=-1, shrinking=True,
             tol=0.001, verbose=False)
new_svr.fit(x_train, y_train)
```

```python
y_pred=new_svr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, r2_score
score=r2_score(y_test,y_pred)
```

```python
score
```
